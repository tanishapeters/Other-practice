```python
#working
#Given an index k, return the kth row of the Pascal’s triangle.

def getRow(k):
    if k<0:
        return [k]  
    result=[[1]]
    for r in range (1, k+1):
        row=[1]
        for c in range (1,r):  
            row.append(result[r-1][c-1] + result[r-1][c]) 
        row.append(1)
        result.append(row)
        
    return result[-1]

getRow(2) 
```

```python
#working
#Pascal’s triangle : To generate A[C] in row R, sum up A’[C] and A’[C-1] from previous row R - 1.
def PascalTriangle(A):
    if A<=0:
        return A
    result=[[1]]
    for r in range (1, A):
        row=[1]
        for c in range (1,r):  
            row.append(result[r-1][c-1] + result[r-1][c]) 
        row.append(1)
        result.append(row)
        
    return result

PascalTriangle(2) 
```

```python
#working
def wave():
    
    A=[5, 1, 3, 2, 4 ]
    A.sort()
    n=len(A)
    
    for i in range (1, n):
        
        if (i%2==0):
            continue
        elif (A[i]>A[i-1]):
            
            A[i-1], A[i]=A[i], A[i-1]
            
        elif (A[i]<A[i-1]):
            continue
            
    return A

wave()
```

```python
#working spiral order another way
class Solution:
    def spiralOrder(self, A):
        l=0
        r=len(A[0])-1
        up=0
        down=len(A)-1 
        dir=0
        array=[]
        while (l<=r and up<=down):
            if (dir==0):
                for c in range (l,r+1):
                    array.append(A[up][c])
                up+=1
            elif dir==1:
                for c in range (up,down+1):
                    array.append(A[c][r])
                r-=1
            elif dir ==2:
                for c in range (r,l-1,-1): #decrementing
                    array.append(A[down][c])
                down-=1
            elif dir ==3:
                for c in range (down,up-1,-1):
                    array.append(A[c][l])
                l+=1
            dir=(dir+1)%4
        return array
```
